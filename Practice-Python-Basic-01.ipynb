```python
# Input a string and print its length.
s = input("Enter the string :")

length = len(s)

print("Lenght is :", length)
```

```python
# Input a string and count vowels.
s = input("Enter string :")
cnt = 0
for i in s:
  if(i=='a' or i=='e' or i=='i' or i=='o' or i=='u' or i=='A' or i=='E' or i=='I' or i=='O' or i=='U'):
    cnt = cnt + 1

print(cnt, "vowels are present")
```

```python
#Input a string and reverse it.
s = input("Enter the string :")
print(s[::-1])
```

```python
#Check whether a string is a palindrome.

s = input("Enter the palindrome :")
size = len(s)

if(s == s[::-1]):
  print("Palindrome")
else:
  print("Not Palindrome")




#string[start:end:step]
#s[start:end:-1]
#start is omitted → start from the end.
#end is omitted → go until the beginning.
#step = -1 → move backwards.


```

```python
#Sort a list in ascending order.
li = [3,2,14,5,3,]
li.sort()
print(li)
```

```python
#Count how many times a number appears in a list.
li = [3,2,14,5,3,]
num = int(input("Enter the number :"))

li = li.count(num)
print(li)
```

```python
#Remove a specific element from a list.

li = [3,2,14,5,3]
li.remove(3) #removes only the first occurrence of the specified value.
print(li)


#to remove all occurrences of 3

li = [3,2,14,5,3]

while 3 in li:
    li.remove(3)

print(li)
```

```python
#Insert an element at a given position.

#ist.insert(index, element)

li = [3,2,14,5,3]
li.insert(2,11)

print(li)
```

```python
#Find the index of an element.

li = [3,2,14,5,3]

indx = li.index(14)
print(indx)

```
