```python
#question 2

#create a 5x5 array(stock level)
stock = np.random.randint(20,80, (5,5))
print( "Available stock : ")
print(stock, '\n')


#create another 5x5 array(demand)

demand = np.random.randint(20,80,(5,5))
print("Demand :")
print(demand, '\n')

#compute : Remaining stock = stock- demand

remain_stock = stock - demand
print("remained_stock :")
print(remain_stock, '\n')


#Replace : Negative value ->0 (out of zero)

replace = np.where(remain_stock<0,0,remain_stock)
print("Negative Replaced with zero:")
print(replace, '\n')


#Find : Total Shortage

total_shortage = np.sum(np.where(remain_stock<0, remain_stock * -1, 0 ))
print("Total Shortage :")
print(total_shortage, '\n')

#products with zero stock

product_with_zero_stock = np.sum(np.sum(np.where(replace ==0, 1, 0) ,axis = 0))
print("product_with_zero_stock :")
print(product_with_zero_stock, '\n')


#sort row based on remaining

sorted = np.sort(replace, axis = 1)
print("sorted :")
print(sorted, '\n')



```
