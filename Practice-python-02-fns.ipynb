Part A: Functions

```python
#What is a function in Python?
#A function is a block of code that performs a specific task and can be reused whenever needed
```

```python
#Why do we use functions?
#Functions help us organize, reuse, and simplify our code.
```

```python
#What is the syntax of a function?

def function_name(parameters):
    # code block
    return value
```

```python
#What is the purpose of the return statement?
#The return statement is used to send a value back from a function to the place where the function was called.
```

```python
#What happens if a function has no return statement?
#If a function does not have a return statement, Python automatically returns None.
```

```python
#What is the difference between parameters and arguments?
# parameter : Written inside the function declaration, Receive data
# Argument : Written inside the function call, Provide data


def square(n):   # n is a parameter
    return n * n

square(5)        # 5 is an argument
```

```python
#Can a function return multiple values?
#Yes! A Python function can return multiple values.

def calculate(a, b):
    return a + b, a - b
```

```python
#Write a function to add two numbers.

num1 = int(input("Enter 1st number :"))
num2 = int(input("Enter 2nd number :"))

def calcu(a,b):
  return a+b, a-b

print(calcu(num1,num2))
```

```python
#Write a function to calculate factorial of a number.
def factorial(n):
    fact = 1

    for i in range(1, n + 1):
        fact *= i

    return fact

num = int(input("Enter a number: "))

print("Factorial =", factorial(num))
```

Part B: Lambda Functions

```python
#What is a lambda function?

#A lambda function is a small anonymous (nameless) function in Python.
#It is used when you need a simple function for a short period of time.



square = lambda x: x * x
print(square(5))
```

```python
#Why is lambda called an anonymous function?
# A lambda function is called an anonymous function because it is created without a name.
```

```python
#syntax:
#syntax :
lambda arguments : expression
```

```python
#Can a lambda function contain multiple statements?
#No. A lambda function cannot contain multiple statements.
```

```python
#Create a lambda function to multiply three numbers.
multiply = lambda a,b,c : a*b*c

print(multiply(10,20,30))
```

```python
#Create a lambda function to find the larger of two numbers.
larger = lambda a,b : a if a>b  else b

print(larger(10,20))
```

Part C: Strings

```python
#Are strings mutable or immutable?
#Strings in Python are immutable.This means that once a string is created, its characters cannot be changed.
#Because strings are immutable, so you cannot modify individual characters.
```

```python
#Difference between upper() and lower().
# upper -> converts all letters to upper, Returns a new string, Does not modify the original string
#lower -> converts all letters to lower, Returns a new string, Does not modify the original string

```

```python
#Difference between capitalize() and title().
# capitalise -> Makes only the first character of the entire string uppercase
# title -> Makes the first character of every word uppercase
```

```python
s = "hello world"

print(s.capitalize())
```

```python
s = "hello world"

print(s.title())
```

```python
#What does find() return?
#he find() method returns the index of the first occurrence of a substring in a string.If the substring is not found, it returns -1.

s = "Hello World"

print(s.find("World"))
```

```python
s = "Hello World"

print(s.find("Python"))
```

```python
#What does count() return?
#The count() method returns the number of times a specified value appears in a string, list, or tuple.

s = "banana"

print(s.count("a"))
```

```python
#What is the use of strip()?
#The strip() method is used to remove spaces (or specified characters) from the beginning and end of a string.

s = "   Hello World   "

print(s.strip())
```

```python
#What is the use of replace()?

#The replace() method is used to replace one substring (or character) with another substring (or character) in a string.

# string.replace(old, new)

s = "I like Python"

print(s.replace("Python", "Java"))
```

```python
#What is the difference between startswith() and endswith()?

# startwith -> Checks whether a string starts with a specified value
# endwith -> Checks whether a string ends with a specified value

#both return true or false

s = "samar kumar"
print(s.startswith("s"))
print(s.endswith("b"))
```

```python
#Count vowels in a string.
s = input("Enter a string: ")
count = 0

for ch in s:
  if ch.lower() in "aeiou":
    count += 1

print("Number of vowels =", count)
```
