```python
a = np.random.randint(1,7,4)  # first argument -> from where to start, 2nd -> end -1, 3rd -> size of array
print(a)

a = np.random.randint(1,7,10)
print(a)

a = np.random.randint(1,7,(4,3))
print(a)
```

```python
a = np.random.choice([2,4,2,9],5)    # this will create an array of size 5, any elements randomly pick from given array to create new array of size 5
print(a)

a = np.random.choice([2,4,2,9],(4,5))
print(a)
```
