```python

scores = []

days = ['monday','tesday','wed','thursday','friday','sat','sun']
for i in range (7):
  x = int(input(f'give the input for the {days[i]}'))         
  scores.append(x)


# scores = input().split()  # method to take input in one line 
# print(scores)
#scores = [int(i) for i in scores]  #this converts string into integer

# print(scores)

a = np.sum(scores)
b = np.mean(scores)
c = np.std(scores)
d = np.min(scores)
e = np.max(scores)
f = np.unique(scores)

print("Total :", a)
print("Mean :",b)
print("Standard Deviation :",c)
print("Minimum :",d)
print("Maximum :",e)
print("Unique :",f)
```
