```python
df = pd.read_csv('student_data.csv')
df.head()
```

```python
sns.lineplot(data=df,x='study_hours',y='exam_score')
plt.tight_layout()
plt.show()
```

```python
def grade(score):
  if score>=90:
    return 'A'
  elif score>=80:
    return 'B'
  elif score>=70:
    return 'C'
  elif score>=60:
    return 'D'
  else:
    return 'F'

df['grade'] = df['exam_score'].apply(grade)
df

```

```python
sns.barplot(data=df,x='grade',y='exam_score',estimator="mean")
plt.title("Average exam score by grade")
plt.tight_layout()
plt.show()
```
