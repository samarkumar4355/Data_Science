```python
s=input()
i=0
j=len(s)-1

is_palindrome = True
while i<j:
  if s[i]!=s[j]:
    is_palindrome = False
    break
  else :
    i+=1
    j-=1
if is_palindrome :
  print("Palindrome")
else :
  print("Not Palindrome")
```

```python
lst = [1, 2, 3, 'hello', 'c', [2, 4]]

print(lst)

# append()
lst.append([2, 3, 4])
print(lst)

#extend()
lst.extend([2, 3, 4])
print(lst)

# insert()

lst.insert(2, "value")
print(lst)

# pop()
lst.pop()
print(lst)

# remove()

lst.remove(2)
print(lst)

# index()

print(lst.index(2))

# count()
print(lst.count(3))

# sort()

lst.sort()

# sum, min, max, sorted
```

```python
lst1 = []
lst2 = []

size = int(input())

for i in range(size):
  x = int(input())

  lst1.append(x)

for j in range(size):
  y = int(input())

  lst2.append(y)

lst3 = []

for i in range(size):
  lst3.append(lst1[i] + lst2[i] )

print(lst3)
```
